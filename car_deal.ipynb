

The dataset is [The Car Evaluation Data Set  from UCI Repository](http://archive.ics.uci.edu/ml/datasets/Car+Evaluation)


The attributes:

1. buying       
2. maint       
3. doors        
4. persons  
5. lug_boot      
6. safety       

Class Values: 

unacc, acc, good, vgood 

### Read data

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# import dataset
column_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety','classify']


raw_data = pd.read_csv("car.data", names=column_names)
data = raw_data.copy()
raw_data.head(20)
```

### Deal data

```python
#转换buying列，将属性值数值化vhigh, high, med, low，分别为0，1，2，3
buying=list(data['buying'])   #找到原表这一列转化为list
for i in buying:  #buying: vhigh, high, med, low
    if 'vhigh'  in i:
        data['buying']=data['buying'].replace(i,'0')
    if 'high'  in i:
        data['buying']=data['buying'].replace(i,'1')
    if 'med'  in i:
        data['buying']=data['buying'].replace(i,'2')
    if 'low'  in i:
        data['buying']=data['buying'].replace(i,'3')
print('buying transform success!')
```

```python
#转化maint列，将属性值vhigh, high, med, low，分别为0，1，2，3
maint=list(data['maint'])   #找到原表这一列转化为list
for i in maint:  #maint: vhigh, high, med, low
    if 'vhigh'  in i:
        data['maint']=data['maint'].replace(i,'0')
    if 'high'  in i:
        data['maint']=data['maint'].replace(i,'1')
    if 'med'  in i:
        data['maint']=data['maint'].replace(i,'2')
    if 'low'  in i:
        data['maint']=data['maint'].replace(i,'3')
print('maint transform success!')
```

```python
#转化doors列，将5more属性值转换为5进行处理
doors=list(data['doors'])   #找到原表这一列转化为list
for i in doors:  #maint: vhigh, high, med, low
    if '5more'  in i:
        data['doors']=data['doors'].replace(i,'5')
print('doors transform success!')
```

```python
#转化lug_boot列，将small, med, big属性值转换为0，1，2处理
lug_boot=list(data['lug_boot'])   #找到原表这一列转化为list
for i in lug_boot:  #lug_boot: small, med, big. 
    if 'small'  in i:
        data['lug_boot']=data['lug_boot'].replace(i,'0')
    if 'med'  in i:
        data['lug_boot']=data['lug_boot'].replace(i,'1')
    if 'big'  in i:
        data['lug_boot']=data['lug_boot'].replace(i,'2')
print('lug_boot transform success!')
```

```python
safety=list(data['safety'])   #找到原表这一列转化为list
for i in safety:  #safety: low, med, high.  
    if 'low'  in i:
        data['safety']=data['safety'].replace(i,'0')
    if 'med'  in i:
        data['safety']=data['safety'].replace(i,'1')
    if 'high'  in i:
        data['safety']=data['safety'].replace(i,'2')
print('safety transform success!')
```

```python
classify=list(data['classify'])   #找到原表这一列转化为list
for i in range(0,len(classify)):  #classify: unacc, acc, good, vgood 
    if 'good' == classify[i]:   
        data['classify'][i] = '2'
    elif 'vgood' == classify[i]:
        data['classify'][i] = '3'
    elif 'unacc' == classify[i]:
        data['classify'][i] = '0'
    elif 'acc' == classify[i]:
        data['classify'][i] = '1'
print('classify transform success!')
```

```python
#说明：more属性在此数值化为5
persons=list(data['persons'])   #找到原表这一列转化为list
for i in persons:  #maint: vhigh, high, med, low
    if 'more'  in i:
        data['persons']=data['persons'].replace(i,'5')
print('persons transform success!')
```
